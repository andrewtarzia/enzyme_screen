## Password for BRENDA is collected from directory on my machine

```python
def get_pw(directory, file):
    """
    Collect password string for BRENDA from directory/file
    """
    with open(directory+file, 'r') as f:
        for line in f:
            l = line.rstrip()
            return l    
```

```python
from requests import Session
from requests.auth import HTTPBasicAuth  # or HTTPDigestAuth, or OAuth1, etc.
from zeep import Client
from zeep.cache import SqliteCache
from zeep.transports import Transport

session = Session()
session.auth = HTTPBasicAuth(username=username, password=get_pw(pw_dir, pw_file))
client = Client(wsdl, transport=Transport(session=session, 
                                          cache=SqliteCache(path='/home/atarzia/Downloads/sqlite.db', timeout=1200)))
#print(client.service)
```
